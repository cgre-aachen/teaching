## Some simple examples on the curse of dimensionaltiy

```python
# basic imports
import numpy as np                                        
import matplotlib.pyplot as plt     
import pickle
import pandas as pd
import scipy.special
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual
from IPython.display import display
```

### Volume of a ball with radius $R$

What happens to the volume of a ball with a certain radius? From the first dimensions, we know that the volume increases:



```python
radius = 1.0
V1 = 2*radius
V2 = np.pi * radius**2
V3 = 4/3. * np.pi * radius**3
```

```python
plt.plot([1,2,3],[V1, V2, V3], 'ro-')
plt.xlabel("Dimension")
plt.ylabel("Volume of sphere with radius $R$")
```

So, what does this mean and how does it continue to higher dimensions?

An asymptotic formula can be obtained using Stirling's approximation:

$V_n(R) \approx \frac{1}{\sqrt{n \pi}} \left( \frac{2 \pi e}{n} \right)^\frac{n}{2} \, R^n$

Try it out yourself:

```python
def v_approx(n, radius=1.0):
    """Approximate volume of sphere with radius r in dimension n"""
    return np.pi**(n/2) / scipy.special.gamma(n/2 + 1.) * radius**n
```

```python
dims = np.arange(20)+1
plt.plot(dims, v_approx(dims), 'o-')
plt.xlabel("Dimensions")
plt.ylabel("Volume of sphere")
plt.grid()
```

### Ratio between volume in hypersphere and hypercube

```python
def ratio(max_dim, radius=1.0):
    dims = np.arange(max_dim)+1
    plt.plot(dims, v_approx(dims) / (2*radius)**dims, 'o-')
    plt.plot(dims, v_approx(dims), 'o-')
    # plt.plot(dims, (2*radius)**dims, 'o-')
    plt.xlabel("Dimensions")
    plt.ylabel("Ratio of sphere tp hypercube")
```

```python
interact(ratio, max_dim=widgets.IntSlider(min=2,max=30,step=1,value=2),
        radius=widgets.FloatSlider(min=0.,max=3.0,step=0.1,value=1.0))
```

### Area of a high-dimensional ball

```python
def area_approx(n, radius=1.):
    """Apprixmation of the area of a sphere with radius r in dimension n"""
    return 2 * np.pi**(n+1/2.) / scipy.special.gamma(n+1 / 2.) * radius**n
```

```python
plt.plot(dims, area_approx(dims), 'o-')
plt.xlabel("Dimensions")
plt.ylabel("Area of sphere")
```

### Ratio of volumes for spheres with different radius



```python
dims = np.arange(20)+1
def vol_ratio(max_dim):
    dims = np.arange(max_dim)+1
    plt.plot(dims, v_approx(dims, radius=1.0)/v_approx(dims, radius=0.9), 'o-')
    plt.xlabel("Dimensions")
    plt.xlim([0,30])
    #plt.ylim([0,25])
    plt.ylabel("Volume of sphere")
    
interact(vol_ratio, max_dim=widgets.IntSlider(min=2,max=30,step=1,value=2))


```
